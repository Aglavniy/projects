## Для каждого вида доп. продукции нужно по посчитать – сколько в среднем заказывают данного вида продукции на один основной товар (только среди заказов, содержащих данный вид доп.продукции). Например, для доп.товара «Вилка» мы должны взять все заказы, содержащие вилку, и поделить количество вилок на количество основных товаров в заказе.

В CSV-файлах представлены два датасета:

Список заказов интернет-магазина. Содержит следующие поля:

*	Уникальный ID заказа
*	ID товара в заказе
*	Заказанное количество данного товара
*	Уникальный ID покупателя
*	Признак дополнительного товара


Cправочник товаров

Товары в заказах делятся на основные (is_additional=0) и «дополнительные» (бесплатная продукция, которая идет к заказу, is_additional=1)
Необходимо получить ответы на следующие вопросы:


```python
# доп. продукция

addit_product = total[total['is_additional'] == 1]
```

```python
# кол-во каждого доп. товара в каждом заказе, где есть доп. товары
addit = addit_product.pivot_table(index='order_id', 
                        columns='title', 
                        values='counter', 
                        aggfunc='sum',
                       fill_value=0)
addit
```

```python
# дф заказов без учета доп. продукций
main_product = total[total['is_additional'] == 0]
```

```python
# кол-во основного товара в каждом заказе

main = main_product.pivot_table(index='order_id', 
                        columns='title', 
                        values='counter', 
                        aggfunc='sum',fill_value=0
                       )

main
```

```python
# считаем кол-во основных товаров по каждому заказу
main['total_count_main']=main.sum(axis=1)
```

```python
df1 = addit.merge(main.iloc[:,-1], 
                  left_index =True, 
                  right_index =True, 
                  how='left' )
```

```python
df1
```

```python
def mean_addit(df, column) -> float:
    """ возвращает среднее по двум столбцам.
        Столбец доп. товара и столбец кол-ва основных товаров 
    """
    
    new_df = df[df[column] >0][[column,'total_count_main']]
    return new_df.sum().iloc[0] / new_df.sum().iloc[1]
```

```python
answer1 = {}
for col in df1.columns[:-1]:
    num = mean_addit(df1, col)
    answer1[col] = num
```

```python
answer1
```

ниже немного симпатичее и нагляднее сделал, в виде таблицы.

```python
asnwer1_df = pd.DataFrame(answer1, 
             index=['Среднее по доп. товару на 1 основной'])\
.T\
.sort_values('Среднее по доп. товару на 1 основной', 
             ascending=False)
#asnwer1_df['Среднее по доп. товару на 1 основной'] = round(asnwer1_df['Среднее по доп. товару на 1 основной'],3)
asnwer1_df
```

Оказалось 6 товаров в 5 заказах, где были основным товаром наши допы. В данном случае считаю стоит оставить, как есть. Ведь данные позици могут быть основнмы товаром также. (вообще в реальной задаче, я бы также уточнил у руководителя, что с такими данными принято делать)

```python
def exercise2(df, column):
    """
    Возвращает датафрейм с топ10 кол-ом основных товаров по каждому допу.
    """
    
    # Фильтруем доп column по таблице, 
    # затем считаем кол-во каждого товара по всем заказам
    # и после сортировки выводим топ-10
    
    new_df = df[df[column]>0].iloc[:,8:]\
    .sum()\
    .sort_values(ascending=False)\
    .head(10)

    
    return pd.DataFrame(new_df, 
                        columns=['кол-во заказов с '+column.replace('_x','')])\
                        .reset_index()\
                        .rename(columns={'title':'название продукта'})
```

```python
# создаем словарь для доп. товроав и таблиц с топ-10 
answer2 ={}
```

```python
for col in df2.columns[:8]:
    answer2[col.replace('_x','')]=exercise2(df2, col)
```

```python
for i,y in  answer2.items():
    display(i)
    display(y)
    print('------------\n')
```
