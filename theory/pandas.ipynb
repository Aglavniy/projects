прикол, почти угадали нижнюю границу, но здесь мы наугад действовали по кол-ву столбцов. Как иначе?

* В данном случае мы изначально применили фильтрацию 'df.isna().sum() > 0', а так как при фильтрации вывод булевы значения, мы применили эту фильтрацию к изначальному датафрейму 'df.isna().sum()'.
* в итоге не нужен head()!

Возникает вопрос, какие типы у данных колонок и как их заменить?
- щас узнаем

1) У категориальных заменим, к примеру, на самую популчрную категорию
2) у числовых на медиану

- Вообще есть много вариантов замены категориальных значений, вплоть до заглушек, топовых значений, применений ml - задача классифицкации. Также разделение их на подкатегории .

```python
# посмотреть на название стобцов с пропусками
na_col.index
```

```python
# Категориальные столбцы
cat_columns = df.columns[df.dtypes == 'object']

# Числовые столбцы
num_columns = df.columns[df.dtypes != 'object']
```

```python
def difference(df_col, iter_col) -> list:
    """ возвращает значения которые есть в двух списках
    """
    spisok=[]
    for col in iter_col:
        if col in df_col:
            spisok.append(col)
    return spisok
```

```python
cat_na = difference(cat_columns, na_col.index)
```

```python
num_na = difference(na_col.index, num_columns)
```

```python
cat_na
```

```python
num_na
```

```python
df['BsmtCond'].value_counts().index[0]
```

Отлично! Мы разделили столбцы с пропусками на категориальные и числовые.
Заменим пропуски!


```python
def fill_nan(columns, cat=True) -> None:
    """ заполняет пропуски
    """
    for col in columns:
        
        if cat:
        # подсчет кол-ва уникальных значений столбца и вывод верхнего
            val = df[col].value_counts().index[0]
        else:
        # медиана для числовых
            val = df[col].median()

        # замена пропуска
        df[col] = df[col].fillna(val)
```

```python
fill_nan(cat_na)
fill_nan(num_na, cat=False)
```

```python
df.isna().sum()[df.isna().sum() > 0]
```

```python
# когда кол-во пропусков по вашему мнению незначительно, а также не повлияет в худшую сторону их удаление
df = df.dropna()
```

### работа с датами

метод to_datetime() нужен для работы с датой

%d — день месяца (от 01 до 31)

%m — номер месяца (от 01 до 12)

%Y — четырёхзначный номер года (например, 2019)

Z — стандартный разделитель даты и времени

%H — номер часа в 24-часовом формате

%I — номер часа в 12-часовом формате

%M — минуты (от 00 до 59)

%S — секунды (от 00 до 59)
