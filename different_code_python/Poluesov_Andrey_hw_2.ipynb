```python
import requests
from bs4 import BeautifulSoup
import sys
import re
import csv
```

```python
#список топ 15 городов РФ
url = 'https://ru.wikipedia.org/wiki/Список_городов_России_с_населением_более_100_тысяч_жителей'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'lxml')
td_al_center = soup.find_all('td', {'align': 'left'})
city_spisok = []
for quote in td_al_center[:15]:
    city_spisok.append(quote.text)
```

```python
city_spisok
```

```python
#проверка почты
regex = re.compile(r'([A-Za-z0-9]+[.-_])*[A-Za-z0-9]+@[A-Za-z0-9-]+(\.[A-Z|a-z]{2,})+')

def check_mail(email):
    if re.fullmatch(regex, email):
        return email
    else:
        wrong_spisok.append('Почта введена неверно')
```

```python
#проверка возраста
def check_age(age):
    if int(age) < 18 or int(age) >65:
        wrong_spisok.append('Возраст не соответсвует критериям')
        
    return age
```

```python
#проверка города
def check_city(city):
    for i in city_spisok:
        if i == city:
            return i
    for i in city_spisok:
        if (len(set(i.lower())&set(city.lower())) +1 == len(set(i.lower()))) &(0 <= (len(i) - len(city)) <= 1):
            return i
    if city not in city_spisok:
        print("Вы не можете участвовать в конференции")
        wrong_spisok.append('Город не из топ-15')
    return city
```

```python
#создаем класс
class Carts:
    #введем списки 
    suit = ['черви','пики','буби','трефы']
    cart = [i for i in range(2,11)] + ['валет','дама','король','туз']
    
    def __init__(self):

        #создадим итераторы для объекта
        self.iter_suit = iter(self.suit)
        self.iter_cart = iter(self.cart)

        #у масти сразу проитерируем
        self.suiter = next(self.iter_suit)

        #счетчик для вызова ошибки
        self.counter =  0
    
    def __iter__(self):
        return self

    def __next__(self):
        # проверка счетчика
        if self.counter == 4:
            raise(StopIteration)    
            
        y = next(self.iter_cart)
        x = self.suiter

        #проверка итератора для обновления итератора и счетчика
        if self.cart[-1] == y:
            if x != self.suit[-1]:

                self.suiter = next(self.iter_suit)
            else:
                self.iter_suit = iter(self.suit)
            self.iter_cart = iter(self.cart)
            self.counter +=1
        
        return f'Номинал: {y}. Масть: {x}'
```

```python
#создаем объект класса
my_class = Carts()
#my_iter = iter(my_class)

while True:
    print(next(my_class))
```
